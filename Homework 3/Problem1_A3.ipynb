# Problem 1. Linear transformation and affine transformation.

## Note: Each column of pos_M is corresponding to a data point

```python
import numpy as np
import matplotlib.pyplot as plt

pos_M = np.array([[-10, -8, -9, -8, -7, -6, -5, -4, -5, -3, -4, -3, -2, -1, 0, -1, 1,0, 1, 2, 3, 4, 3, 5, 4, 5, 6, 7, 8, 9, 8, 10],
                  [-1, -1, -1, 0, 1, 2, 3, 4, 4, 4, 4, 3, 2, 1, 0, 0, 0, 0, 1, 2, 3, 4, 4, 4, 4, 3, 2, 1, 0, -1, -1, -1]])
# each column corresponding to a data point

print("Shape of pos_M is {}".format(pos_M.shape))

def plot_shape(P):
  # P is a 2-by-N matrix for
  plt.figure()
  plt.plot(P[0], P[1])
  ax = plt.gca()
  #ax.axes.xaxis.set_visible(False)
  #ax.axes.yaxis.set_visible(False)
  ax.set_aspect('equal')
  plt.show()


plot_shape(pos_M)

def plot_resulting_shape(pos_M_result, pos_M):
  plt.figure()
  plt.plot(pos_M[0], pos_M[1])
  plt.plot(pos_M_result[0], pos_M_result[1])
  ax = plt.gca()
  ax.axes.xaxis.set_visible(False)
  ax.axes.yaxis.set_visible(False)
  ax.set_aspect('equal') # Updated here
  plt.show()


```

##Problem 1.1 Single transformation



For each prescribed transformation below, please


1.  construct the transformation matrix;
2.  apply them on pos_M;
3.  call the plot function to plot the resulting shapes.

```python
# Scaling
# Please scale the shape so that it has
# twice the current width in the x direction
# half the current height in the y direction
print("Scaling result:")
# Scaling transformation matrix
A_scale = np.array([[2, 0],
                    [0, 0.5]])

# Apply scaling transformation to the base "M" shape
pos_M_scale = A_scale @ pos_M

# Plot the resulting shape
plot_resulting_shape(pos_M_scale, pos_M)
```

```python
# Rotation
# Please rotate the shape counterclockwise by 60 degrees
print("Rotation result:")
# Rotation transformation matrix
theta = np.pi / 3  # 60 degrees in radians
A_rotate = np.array([[np.cos(theta), -np.sin(theta)],
                     [np.sin(theta), np.cos(theta)]])

# Apply rotation transformation to the base "M" shape
pos_M_rotate = A_rotate @ pos_M

# Plot the resulting shape
plot_resulting_shape(pos_M_rotate, pos_M)
```

```python
# Translation
# Please translate the shape along x axis with length 10
# and along y axis with length 20
print("Translation result:")
# Translation transformation matrix
tx = 10
ty = 20
A_translate = np.array([[1, 0],
                        [0, 1]]) + np.array([[tx], [ty]])

# Apply translation transformation to the base "M" shape
pos_M_translate = A_translate @ pos_M

# Plot the resulting shape
plot_resulting_shape(pos_M_translate, pos_M)
```

```python
# Shearing 1
# Please shear the shape along the x-direction with a shearing factor 3
# and keep the shape along the y-direction
print("Shearing 1 result:")
# Shearing transformation matrix
sx = 3
sy = 0
A_shear_h = np.array([[1, sx],
                      [sy, 1]])

# Apply shearing transformation to the base "M" shape
pos_M_shear_h = A_shear_h @ pos_M

# Plot the resulting shape
plot_resulting_shape(pos_M_shear_h, pos_M)
```

```python
# Shearing 2
# Please shear the shape along the y-direction with a shearing factor 6
# and keep the shape along the x-direction
print("Shearing 2 result:")
# Shearing transformation matrix
sx = 0
sy = 6
A_shear_v = np.array([[1, sx],
                      [sy, 1]])

# Apply shearing transformation to the base "M" shape
pos_M_shear_v = A_shear_v @ pos_M

# Plot the resulting shape
plot_resulting_shape(pos_M_shear_v, pos_M)
```

## Problem 1.2 Compositions of  Transformations

### Please use the **transformation matrices constructed in Problem 1.1** to apply multiple transformations to the base shape then call the plot function to plot the resulting shapes.


```python
# Sequence 1
# Please do the following two things in sequence
# 1. scale the shape so that it has
#    * twice the current width in the x direction
#    * half the current height in the y direction
# 2. rotate the resulting shape counterclockwise by 60 degrees
print("Sequence 1 result: ")
# Step 1: Scaling transformation matrix
A_scale_seq1 = np.array([[2, 0],
                         [0, 0.5]])

# Step 2: Rotation transformation matrix
theta_seq1 = np.pi / 3  # 60 degrees in radians
A_rotate_seq1 = np.array([[np.cos(theta_seq1), -np.sin(theta_seq1)],
                           [np.sin(theta_seq1), np.cos(theta_seq1)]])

# Composition of transformations: Rotate * Scale
A_seq1 = A_rotate_seq1 @ A_scale_seq1

# Apply the composed transformation to the base "M" shape
pos_M_seq1 = A_seq1 @ pos_M

# Plot the resulting shape
plot_resulting_shape(pos_M_seq1, pos_M)
```

```python
# Sequence 2
# Please do the following two things in sequence
# 1. rotate the shape counterclockwise by 60 degrees
# 2. scale the resulting shape so that it has
#    * twice the current width in the x direction
#    * half the current height in the y direction
print("Sequence 2 result: ")
# Step 1: Rotation transformation matrix
theta_seq2 = np.pi / 3  # 60 degrees in radians
A_rotate_seq2 = np.array([[np.cos(theta_seq2), -np.sin(theta_seq2)],
                           [np.sin(theta_seq2), np.cos(theta_seq2)]])

# Step 2: Scaling transformation matrix
A_scale_seq2 = np.array([[2, 0],
                         [0, 0.5]])

# Composition of transformations: Scale * Rotate
A_seq2 = A_scale_seq2 @ A_rotate_seq2

# Apply the composed transformation to the base "M" shape
pos_M_seq2 = A_seq2 @ pos_M

# Plot the resulting shape
plot_resulting_shape(pos_M_seq2, pos_M)
```

```python
# Sequence 3
# Please do the following three things in sequence
# 1. rotate the shape counterclockwise by 60 degrees
# 2. shear the resulting shape along the y-direction with a shearing factor 6
#      while keeping the shape along the x-direction
# 3. scale the resulting shape so that it has
#    * twice the current width in the x direction
#    * half the current height in the y direction
print("Sequence 3 result: ")
# Step 1: Rotation transformation matrix
theta_seq3 = np.pi / 3  # 60 degrees in radians
A_rotate_seq3 = np.array([[np.cos(theta_seq3), -np.sin(theta_seq3)],
                           [np.sin(theta_seq3), np.cos(theta_seq3)]])

# Step 2: Shearing transformation matrix along the y-direction
sx_seq3 = 0
sy_seq3 = 6
A_shear_y_seq3 = np.array([[1, sx_seq3],
                            [sy_seq3, 1]])

# Step 3: Scaling transformation matrix
A_scale_seq3 = np.array([[2, 0],
                         [0, 0.5]])

# Composition of transformations: Scale * Shear * Rotate
A_seq3 = A_scale_seq3 @ A_shear_y_seq3 @ A_rotate_seq3

# Apply the composed transformation to the base "M" shape
pos_M_seq3 = A_seq3 @ pos_M

# Plot the resulting shape
plot_resulting_shape(pos_M_seq3, pos_M)
```

```python
# Sequence 4
# Please do the following three things in sequence
# 1. shear the shape along the y-direction with a shearing factor 6
#     while keeping the shape along the x-direction;
# 2. rotate the shape counterclockwise by 60 degrees;
# 3.  scale the resulting shape so that it has
#     * twice the current width in the x direction
#     * half the current height in the y direction
print("Sequence 4 result: ")
# Step 1: Shearing transformation matrix along the y-direction
sx_seq4 = 0
sy_seq4 = 6
A_shear_y_seq4 = np.array([[1, sx_seq4],
                            [sy_seq4, 1]])

# Step 2: Rotation transformation matrix
theta_seq4 = np.pi / 3  # 60 degrees in radians
A_rotate_seq4 = np.array([[np.cos(theta_seq4), -np.sin(theta_seq4)],
                           [np.sin(theta_seq4), np.cos(theta_seq4)]])

# Step 3: Scaling transformation matrix
A_scale_seq4 = np.array([[2, 0],
                         [0, 0.5]])

# Composition of transformations: Scale * Rotate * Shear
A_seq4 = A_scale_seq4 @ A_rotate_seq4 @ A_shear_y_seq4

# Apply the composed transformation to the base "M" shape
pos_M_seq4 = A_seq4 @ pos_M

# Plot the resulting shape
plot_resulting_shape(pos_M_seq4, pos_M)
```

## 1.3 Inverse transformation

### Hint: undoing a transformation is performing the **inverse** transformation

```python
# Please undo the rotation in Problem 1.1, i.e.,
# rotate the resulting shape pos_M_rotate there clockwise by 60 degrees
print("Undo Rotation Result: ")
# Step 1: Find the inverse of rotation transformation
theta_inverse_rotate = -np.pi / 3  # Clockwise rotation by 60 degrees
A_inverse_rotate = np.array([[np.cos(theta_inverse_rotate), -np.sin(theta_inverse_rotate)],
                              [np.sin(theta_inverse_rotate), np.cos(theta_inverse_rotate)]])

# Step 2: Apply the inverse rotation transformation to pos_M_rotate
pos_M_rotate_back = A_inverse_rotate @ pos_M_rotate

# Plot the resulting shape
plot_shape(pos_M_rotate_back)
```

```python
# Please undo the shearing 1 of Problem 1.1, i.e.,
# tranform the pos_M_shear_h in Problem 1.1 back to pos_M (original shape)
print("Undo Shearing 1 Result: ")
# Step 1: Find the inverse of shearing 1 transformation
A_inverse_shear_h = np.array([[1, -0.5],
                               [0, 1]])

# Step 2: Apply the inverse shearing 1 transformation to pos_M_shear_h
pos_M_shear_h_back = A_inverse_shear_h @ pos_M_shear_h

# Plot the resulting shape
plot_shape(pos_M_shear_h_back)
```

```python
# Please find the inverse of Sequence 1 in Problem 1.2
# and apply it to pos_M_seq1
print("Undo Sequence 1 Result: ")
# Step 1: Find the inverse of rotation transformation
A_inv_rotate_seq1 = np.linalg.inv(A_rotate_seq1)

# Step 2: Find the inverse of scaling transformation
A_inv_scale_seq1 = np.linalg.inv(A_scale_seq1)

# Composition of inverse transformations: Inverse Rotate * Inverse Scale
A_inv_seq1 = A_inv_scale_seq1 @ A_inv_rotate_seq1

# Apply the composed inverse transformation to pos_M_seq1
pos_M_seq1_back = A_inv_seq1 @ pos_M_seq1

# Plot the resulting shape
plot_shape(pos_M_seq1_back)
```

```python
# Please find the inverse of Sequence 2 in Problem 1.2
# and apply it to pos_M_seq2
print("Undo Sequence 2 Result: ")
# Step 1: Find the inverse of scaling transformation
A_inv_scale_seq2 = np.linalg.inv(A_scale_seq2)

# Step 2: Find the inverse of rotation transformation
A_inv_rotate_seq2 = np.linalg.inv(A_rotate_seq2)

# Composition of inverse transformations: Inverse Scale * Inverse Rotate
A_inv_seq2 = A_inv_rotate_seq2 @ A_inv_scale_seq2

# Apply the composed inverse transformation to pos_M_seq2
pos_M_seq2_back = A_inv_seq2 @ pos_M_seq2

# Plot the resulting shape
plot_shape(pos_M_seq2_back)
```
